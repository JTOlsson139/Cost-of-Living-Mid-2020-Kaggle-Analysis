```python
# Data Analysis and Visualization
# First import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
# Second load data
data_2021 = pd.read_csv('Data\datafile2021.csv')
data_2019 = pd.read_csv('Data\datafile2019.csv')
top50_2021 = data_2021.head(50)
top50_2019 = data_2019.head(50)
```

```python
# Third load Matplotlib Notebook
%matplotlib notebook
```

```python
# QUESTION 1: Did the Countries ranked Top 50 change in their Cost of Living Index?
# First I created a line plot with the ranked Top 50 Countries of 2019.

# List x and y-axis data
country19 = top50_2019.iloc[:,1]
cl_index_2019 = top50_2019.iloc[:,3]
country21 = top50_2021.iloc[:,1]
cl_index_2021 = top50_2021.iloc[:,3]

# Plot the charts and apply some styling
country_2019, = plt.plot(country19, cl_index_2019, color="red", label="2019")
country_2021, = plt.plot(country19, cl_index_2021, color="blue", label="2021")

# Add labels to X and Y axes :: Add title
plt.title("Countries Cost of Living Index 2019 and 2021")
plt.xlabel("Country")
plt.ylabel("Cost of Living Index")

# Rotating X-axis labels
# LINK: https://www.geeksforgeeks.org/how-to-rotate-x-axis-tick-label-text-in-matplotlib/
plt.xticks(rotation = 90)

# Create a legend for the chart
plt.legend(handles=[country_2019, country_2021], loc="best")

# Add in a grid for the chart
plt.grid()
plt.show()
plt.tight_layout()

# Save the image and show the plot
plt.savefig("Images/Image1.png")
plt.show()
```

```python
# Second, I created another line plot, this time using the ranked Top 50 Countries in 2021
country19 = top50_2019.iloc[:,1]
cl_index_2019 = top50_2019.iloc[:,3]
country21 = top50_2021.iloc[:,1]
cl_index_2021 = top50_2021.iloc[:,3]


# Plot the charts and apply some styling
country_2019, = plt.plot(country21, cl_index_2019, color="red", label="2019")
country_2021, = plt.plot(country21, cl_index_2021, color="blue", label="2021")

# Add labels to X and Y axes :: Add title
plt.title("Countries Cost of Living Index 2019 and 2021")
plt.xlabel("Country")
plt.ylabel("Cost of Living Index")

# Rotating X-axis labels
plt.xticks(rotation = 90)

# Create a legend for the chart
plt.legend(handles=[country_2019, country_2021], loc="best")

# Add in a grid for the chart
plt.grid()
plt.show()
plt.tight_layout()

#
plt.savefig("Images/Image2.png")
plt.show()
```

```python
# QUESTION 2: Does the Top 50 median Cost of Living Index value have a greater or smaller correlation to the
# median Rent Index Value in 2019 than 2021?
x = top50_2019["Cost of Living Index"]
y = top50_2019["Rent Index"]
plt.scatter(x, y)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.grid()
plt.savefig("Images/Image3.png")
plt.show()
```
